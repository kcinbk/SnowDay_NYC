```python
import geopandas as gpd
import pandas as pd

# Base URL
base_url = "https://data.cityofnewyork.us/resource/inkn-q76z.geojson"

# Parameters
limit = 50000  # maximum rows per request
offset = 0
all_data = []

while True:
    url = f"{base_url}?$limit={limit}&$offset={offset}"
    gdf = gpd.read_file(url)
    if gdf.empty:
        break
    all_data.append(gdf)
    offset += limit
    print(f"Downloaded {len(gdf)} rows, total so far: {offset}")

# Concatenate all GeoDataFrames
nyc_centerline = pd.concat(all_data, ignore_index=True)
nyc_centerline = gpd.GeoDataFrame(nyc_centerline, geometry='geometry')
print(nyc_centerline.shape)
```
