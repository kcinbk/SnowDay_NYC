```python
from get_opendata import get_data
```

```python
from datetime import datetime

now_datetime_iso =datetime.now().isoformat()
now_datetime_iso
```

```python
data = get_data(
    endpoint = 'data.cityofnewyork.us', 
    dataset_id = 'erm2-nwe9', 
    datetime_field = 'created_date',
    start_date = '2026-02-22T12:00:00', 
    end_date=now_datetime_iso,
    first_filter_field='complaint_type',
    first_filter_value='Snow or Ice',
    limit = 500000         
)
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.shape
```
